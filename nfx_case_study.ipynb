```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("netflix.csv")
df
```

**Handling null values**

From the data we can see that there are multiple null values in the dataset and overcome this, we use fillna() method to fill all the values of the dataframe with some value.

```python
df['cast'] = df['cast'].fillna("unknown character")
df['country'] = df['country'].fillna("unknown country")
df['director'] = df['director'].fillna("unknown director")
df['rating'] = df['rating'].fillna(0)
df['duration'] = df['duration'].fillna(0)
df['date_added'] = df['date_added'].fillna('unknown')
df
```

# Q-3. What is the best time to launch a TV show?

> The problem statement demands the best time to launch a movie using the data provided.

> The solution to the statement can be get through getting the total count of the movies launched each week and each month

> The maximum number of movies released in a perticular week or month is considered the best time to launch a movie.

> The solution has four parts in it.

> one for weekly TV Shows and another weekly movies

> And similarly for monthly TV shows and movies

```python
data.info()
```

```python

# convert date string to week and months
req_df = data[['date_added', 'type', 'title']].copy()
req_df['date_added'] = pd.to_datetime(req_df['date_added'])
req_df['week'] = req_df['date_added'].dt.isocalendar().week
req_df['month'] = req_df['date_added'].dt.month
```

```python
#group data weekly as per requirement for movies

movies_date_df = req_df[req_df['type'] == 'Movie']
week_movie_count = movies_date_df.groupby('week')['title'].nunique().reset_index()
week_movie_count = week_movie_count.sort_values(by='title', ascending=False)
top_5_weeks = week_movie_count.head(5)
top_5_weeks
```

```python
#group data weekly as per requirement for tv shows

tv_date_df = req_df[req_df['type'] == 'TV Show']
week_tv_count = tv_date_df.groupby('week')['title'].nunique().reset_index()
week_tv_count = week_tv_count.sort_values(by='title', ascending=False)
top_10_weeks_tv = week_tv_count.head(10)
top_10_weeks_tv
```

```python
#group data monthly as per requirement for movies

movies_date_df = req_df[req_df['type'] == 'Movie']
month_movie_count = movies_date_df.groupby('month')['title'].nunique().reset_index()
month_movie_count = month_movie_count.sort_values(by='title', ascending=False)
top_10_month = month_movie_count.head(10)
top_10_month
```

```python
#group data monthly as per requirement for tv shows

tv_date_df = req_df[req_df['type'] == 'TV Show']
month_tv_count = tv_date_df.groupby('month')['title'].nunique().reset_index()
month_tv_count = month_tv_count.sort_values(by='title', ascending=False)
top_months_tv = month_tv_count.head(12)
top_months_tv
```

# Q-6. After how many days the movie will be added to Netflix after the release
> For this we need to subtract release date from the date added on netflix.

> Then we will be finding the mean of the difference column.

> The mean will give us the average time a movie will take to come on the platform.

**NOTE:**  

*   The data in the dataset was not sufficient to count the number of days. So no assumptions were made in the above dataset.
*   Some movies were released before 1990s and were available on the platform after 2000s. So, the data might not give the desired result.





```python
required_data = data[data['date_added'] != 0]
required_data = data[['date_added','release_year']].copy()

required_data['date_added'] = pd.to_datetime(required_data['date_added'])
required_data['release_year'] = pd.to_datetime(required_data['release_year'], format='%Y')

# Extract year from the datetime columns
required_data['date_movie_added'] = required_data['date_added'].dt.year
required_data['release_date'] = required_data['release_year'].dt.year
required_data
```

```python
required_data["difference"] = required_data['date_movie_added'] - required_data['release_date']
required_data
```

```python
average_time = required_data['difference'].mean()
average_time
```
